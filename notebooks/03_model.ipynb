# Model training

#### imports & settings

```python
# basics
import numpy as np
import pandas as pd

# model selection
from sklearn.model_selection import train_test_split


# settings
pd.set_option('display.float_format', '{:,.2f}'.format)
random_state = 42
```

#### Read data

```python
df = pd.read_csv("../data/raw/hr-analytics-prediction/HR-Employee-Attrition.csv", 
                 usecols=["Attrition",
                          "Age",
                          "BusinessTravel", 
                          "Education", 
                          "EducationField", 
                          "EnvironmentSatisfaction",
                          "JobInvolvement",
                          "JobLevel",
                          "JobRole",
                          "JobSatisfaction",
                          "MaritalStatus",
                          "OverTime",
                          "RelationshipSatisfaction",
                          "StockOptionLevel",
                          "TotalWorkingYears",
                          "TrainingTimesLastYear",
                          "WorkLifeBalance",
                          "YearsAtCompany",
                          "YearsWithCurrManager"])
df
```

```python
# train-test-split
X_train, X_test, y_train, y_test = train_test_split(df.drop("Attrition", axis=1),
                                                    df["Attrition"],
                                                    test_size=0.2,
                                                    random_state=random_state,
                                                    stratify=df["Attrition"])
```
